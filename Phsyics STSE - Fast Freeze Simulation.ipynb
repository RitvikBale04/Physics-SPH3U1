# Imports

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import style

style.use('ggplot')
```

# Initializing Variable and Arrays

```python
#Initializing starting Temperature to 20 Degrees Celsius
starting_temp = 20

#Number of Simulation
num_simulations = 1000

#Number of Steps
num_steps = 250

#Arbitrary SDs
FastFreeze_SD = 0.05
Freezer_SD = 0.08
Refrigerator_SD = 0.1

#Arbitraty Means
FastFreeze_Mean = -0.2
Freezer_Mean = -0.12
Refrigerator_Mean = -0.08
```

```python
#Initializing the Simulation DataFrame for each Freezing Methody
FastFreeze_simulation_df = pd.DataFrame()

Freezer_simulation_df = pd.DataFrame()

Refrigerator_simulation_df = pd.DataFrame()
```

# Running the Simulations for Each

Fast Freeze

```python
#Running the Simultation for FlashFreeze

for x in range(num_simulations):
    count = 0
    
    price_series = []
    price_series.append(20)
    
    for y in range(num_steps):
        if count == 249:
            break
        price = price_series[count] + (np.random.normal(0, FastFreeze_SD) + FastFreeze_Mean)
        price_series.append(price)
        count += 1
        
    FastFreeze_simulation_df[x] = price_series        
```

Freezer

```python
#Running the Simultation for Freezer

for x in range(num_simulations):
    count = 0
    
    price_series = []
    price_series.append(20)
    
    for y in range(num_steps):
        if count == 249:
            break
        price = price_series[count] + (np.random.normal(0, Freezer_SD) + Freezer_Mean)
        price_series.append(price)
        count += 1
        
    Freezer_simulation_df[x] = price_series        
```

Refrigerator

```python
#Running the Simultation for Refrigerator

for x in range(num_simulations):
    count = 0
    
    price_series = []
    price_series.append(20)
    
    for y in range(num_steps):
        if count == 249:
            break
        price = price_series[count] + (np.random.normal(0, Refrigerator_SD) + Refrigerator_Mean)
        price_series.append(price)
        count += 1
        
    Refrigerator_simulation_df[x] = price_series        
```

```python
FastFreeze_simulation_df.transpose()
Freezer_simulation_df.transpose()
Refrigerator_simulation_df.transpose()
```

# Graphs

Fast Freeze

```python
#Plotting the simulation
fig = plt.figure()
plt.plot(FastFreeze_simulation_df)
plt.axhline(y = 0)
plt.ylim([-40,20])
plt.title('Flash Freeze')
plt.show()

#Initializing an array which will store the time each simulation crosses 0 (Freezing)
time = []

for x in FastFreeze_simulation_df:
    for y in range(num_steps):
        if FastFreeze_simulation_df.loc[y, x] < 0:
            time.append(y)
            break
            
sns.distplot(time)
print(np.mean(time))
print(np.std(time))
```

Freezer

```python
fig = plt.figure()
plt.plot(Freezer_simulation_df)
plt.axhline(y = 0) 
plt.ylim([-40,20])
plt.title('Freezer')
plt.show()

#Initializing an array which will store the time each simulation crosses 0 (Freezing)
time = []

for x in Freezer_simulation_df:
    for y in range(num_steps):
        if Freezer_simulation_df.loc[y, x] < 0:
            time.append(y)
            break
            
sns.distplot(time)
print(np.mean(time))
print(np.std(time))
```

Refrigerator

```python
fig = plt.figure()
plt.plot(Refrigerator_simulation_df)
plt.axhline(y = 0) 
plt.ylim([-40,20])
plt.title('Refrigerator')
plt.show()

#Initializing an array which will store the time each simulation crosses 0 (Freezing)
time = []

for x in Refrigerator_simulation_df:
    for y in range(num_steps):
        if Refrigerator_simulation_df.loc[y, x] < 0:
            time.append(y)
            break
            
sns.distplot(time)
print(np.mean(time))
print(np.std(time))
```
